```python
#ladowanie biblioteki Pandas
import pandas as pd
```

```python
#Zadanie 1
df = pd.read_csv('IHME_ORB_C19HSDS_2020_Y2020M12D03.CSV')
df
print(df.head())

# Sprawd podstawowe informacje o danych
print(df.info())

# Wywietl podstawowe statystyki opisowe
print(df.describe())
```

```python
# Zadanie 2
mean_weight = df["kenya_weight"].mean()
print(f"Srednia waga kenya : {mean_weight}")

# Oblicz median dla kolumny ’dochd’
median_income = df["nigeria_weight"].median()
print(f"Mediana wagi nigeria: {median_income}")

# Oblicz odchylenie standardowe dla kolumny ’kenya_weight’
std_age = df["kenya_weight"].std()
print(f"Odchylenie standardowe wieku: {std_age}")
```

```python
# Sprawdzenie, które kolumny zawierają brakujące wartości
missing_values = df.isnull().sum()
print("Brakujące wartości w każdej kolumnie:")
print(missing_values)

# Uzupełnienie brakujących wartości średnią w kolumnie 'wiek'
df['wiek'].fillna(df['wiek'].mean(), inplace=True)

# Usunięcie wierszy, gdzie brakuje danych w kolumnie 'dochd'
df.dropna(subset=['dochd'], inplace=True)

# Ponowne sprawdzenie brakujących wartości po operacjach
missing_values_after = df.isnull().sum()
print("\nBrakujące wartości po uzupełnieniu i usunięciu:")
print(missing_values_after)
```
