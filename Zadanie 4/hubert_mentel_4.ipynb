```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#WYKRES LINIOWY
# Dane o kartach graficznych i ich cenach
x = ["RTX 4060", "RTX 4070", "RTX 4080", "RTX 4090"]
y = [299, 599, 1199, 1599]  # Ceny w USD

# Tworzenie wykresu
plt.plot(x, y, marker='o', linestyle='-', color='b', label="Ceny GPU")
plt.xlabel("Model karty graficznej (seria 4000)")
plt.ylabel("Cena (USD)")
plt.title("Wzrost cen kart graficznych serii 4000")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#WYKRES SUPKOWY
# Dane o kartach graficznych serii 3000 i 4000 oraz ich cenach
kategorie = [
    "RTX 3060", "RTX 3070", "RTX 3080", "RTX 3090", 
    "RTX 4060", "RTX 4070", "RTX 4080", "RTX 4090"
]
wartosci = [
    329, 499, 699, 1499, 299, 599, 1199, 1599  # Ceny w USD
]

# Tworzenie wykresu supkowego
plt.bar(kategorie, wartosci, color='orange')
plt.xlabel("Model karty graficznej")
plt.ylabel("Cena (USD)")
plt.title("Ceny kart graficznych serii 3000 i 4000")
plt.xticks(rotation=45)
plt.show()
```

```python
#HISTOGRAM
# Rcznie generowane dane dotyczce wydajnoci kart graficznych FPS
# rednia wydajno to 100 FPS, a odchylenie standardowe to 10 FPS
dane = np.random.normal(100, 10, 1000)

# Tworzenie histogramu
plt.hist(dane, bins=30, color='blue', alpha=0.7, density=True)

# Etykiety i tytu wykresu
plt.xlabel("Wydajno (FPS)")
plt.ylabel("Prawdopodobiestwo wydajnoci")
plt.title("Histogram wydajnoci kart graficznych")

# Wywietlenie wykresu
plt.show()
```

```python
#WYKRES KOOWY
# Dane dotyczce sprzeday kart graficznych z serii RTX 4000
modele = ['RTX 4060', 'RTX 4070 Ti', 'RTX 4080', 'RTX 4090']
sprzedaz = [20000, 50000, 30000, 10000]  # Przykadowe dane

# Tworzenie wykresu koowego
plt.pie(sprzedaz, labels=modele, autopct='%1.1f%%', startangle=90, colors=['skyblue', 'orange', 'green', 'purple'])
plt.title("Udzia sprzeday kart graficznych NVIDIA RTX 4000")
plt.show()
```

```python
# KOLORY, LINIE I STYLE MARKERW
# Liczba rdzeni CUDA dla kart RTX 4060, 4070, 4080 i 4090
rdzenie_cuda = [3584, 5888, 8704, 10496]

# Liczba rdzeni ROP dla kart RTX 4060, 4070, 4080 i 4090 (fikcyjne dane)
rdzenie_rop = [48, 64, 80, 96]

# Dostosowany wykres liniowy
plt.plot(rdzenie_cuda, rdzenie_rop, color='red', marker='x', linestyle='--', linewidth=2)

# Etykiety i tytu wykresu
plt.xlabel("Liczba rdzeni CUDA")
plt.ylabel("Liczba rdzeni ROP")
plt.title("Porwnanie rdzeni CUDA i ROP w kartach graficznych RTX 4060, 4070, 4080, 4090")

# Wywietlenie wykresu
plt.show()
```

```python
#DODAWANIE SIATKI
plt.plot(rdzenie_cuda, rdzenie_rop, color='red', marker='o')
plt.grid(True)
plt.xlabel("Liczba rdzeni CUDA")
plt.ylabel("Liczba rdzeni ROP")
plt.title("Wykres z Siatk")
plt.show()
```

```python
#DODAWANIE ADNOTACJI
plt.plot(rdzenie_cuda, rdzenie_rop, color='blue', marker='o')
plt.annotate('Najwiksza wydajno', xy=(10496, 96), xytext=(7000, 90), arrowprops=dict(facecolor='black', arrowstyle='->'))
plt.xlabel("Liczba rdzeni CUDA")
plt.ylabel("Liczba rdzeni ROP")
plt.title("Wykres z adnotacj")
plt.show()
```

```python
#WYKRES PUNKTOWY
# Dane: Wydajno (TFLOPS), liczba rdzeni CUDA i ceny kart graficznych z serii 2000, 3000 i 4000
#"RTX 2060", "RTX 2070", "RTX 2080", "RTX 3060", "RTX 3070", "RTX 3080", "RTX 4060", "RTX 4070", "RTX 4080", "RTX 4090"
wydajnosc_tflops = [6.5, 7.5, 10.0, 13.0, 20.0, 29.8, 20.0, 25.0, 35.0, 40.0]
rdzenie_cuda = [1920, 2304, 2944, 3584, 5888, 8704, 3584, 5888, 8704, 10496]  # Liczba rdzeni CUDA
ceny = [350, 500, 700, 800, 1100, 1500, 600, 800, 1200, 1600]  # Przykadowe ceny w dolarach

sizes = np.array(rdzenie_cuda) / 10  # Mniejszy wspczynnik, aby rozmiar punktw nie by za duy

# Kolory punktw na podstawie ceny
colors = np.array(ceny)

# Tworzenie wykresu punktowego
plt.scatter(wydajnosc_tflops, rdzenie_cuda, s=sizes, c=colors, alpha=0.6, cmap='viridis')

# Dodanie paska kolorw
plt.colorbar(label='Cena (USD)')

# Etykiety i tytu wykresu
plt.xlabel("Wydajno (TFLOPS)")
plt.ylabel("Liczba rdzeni CUDA")
plt.title("Porwnanie kart graficznych RTX 2000, 3000, 4000")

# Wywietlenie wykresu
plt.show()
```

```python
#WYKRES 3D
from mpl_toolkits.mplot3d import Axes3D

# Tworzenie wykresu 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Wykres 3D, gdzie:
# X = wydajno w TFLOPS
# Y = liczba rdzeni CUDA
# Z = cena kart graficznych
ax.scatter(wydajnosc_tflops, rdzenie_cuda, ceny, c=ceny, cmap='viridis', s=100)

# Etykiety i tytu wykresu
ax.set_xlabel('Wydajno (TFLOPS)')
ax.set_ylabel('Liczba rdzeni CUDA')
ax.set_zlabel('Cena (USD)')
ax.set_title('Porwnanie kart graficznych RTX 2000, 3000, 4000')

# Dodanie paska kolorw
fig.colorbar(ax.scatter(wydajnosc_tflops, rdzenie_cuda, ceny, c=ceny, cmap='viridis', s=100), ax=ax, label='Cena (USD)')

# Wywietlenie wykresu
plt.show()
```
