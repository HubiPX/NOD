```python
import numpy as np
import matplotlib.pyplot as plt

# Definicja funkcji Swish
def swish(x):
    return x / (1 + np.exp(-x))

# Pochodna funkcji Swish (gradient)
def grad_swish(x):
    return swish(x) + (1 / (1 + np.exp(-x))) * (1 - swish(x))

# Przygotowanie danych wejściowych
x_vals = np.linspace(-5, 5, 400)

# Obliczanie wartości funkcji Swish oraz jej gradientu
y_vals = swish(x_vals)
grad_vals = grad_swish(x_vals)

# Tworzenie wykresu
plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals, label='Swish(x)', color='orange', linewidth=2)
plt.plot(x_vals, grad_vals, label='Gradient Swish(x)', color='blue', linestyle='--', linewidth=2)
plt.axhline(0, color='black',linewidth=1)
plt.axvline(0, color='black',linewidth=1)
plt.title('Funkcja Swish i jej Gradient')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```
