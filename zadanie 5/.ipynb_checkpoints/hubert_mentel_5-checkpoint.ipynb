```python
import pandas as pd
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
import matplotlib.pyplot as plt
import numpy as np
from sklearn.ensemble import IsolationForest
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Wczytanie pliku CSV
file_path = "gpu_data.csv"  # Upewnij siÄ, Å¼e plik jest w tym samym katalogu co skrypt
data = pd.read_csv(file_path)

# Zmiana ustawieÅ wyÅwietlania, aby pokazaÄ wszystkie wiersze
pd.set_option('display.max_rows', None)

# WyÅwietlenie wszystkich wierszy
print(data)
```

```python
# 1. ZidentyfikowaÄ wartoÅci odstajÄce za pomocÄ algorytmu Isolation Forest
df = pd.read_csv("gpu_data.csv")

# Wybierz tylko kolumny liczbowe do analizy
numerical_df = df.select_dtypes(include=['float64', 'int64'])

# Przeskalowanie danych
scaler = StandardScaler()
numerical_df_scaled = scaler.fit_transform(numerical_df)

# Model Isolation Forest
iso_forest = IsolationForest(contamination=0.05, random_state=42)
outliers = iso_forest.fit_predict(numerical_df_scaled)

# Oznaczenie wartoÅci odstajÄcych (-1 oznacza odstajÄce, 1 oznacza normalne)
df['WartoÅci odstajÄce'] = np.where(outliers == -1, 'OdstajÄce', 'Normal')

# Wizualizacja rozkÅadu wartoÅci odstajÄcych
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='WartoÅci odstajÄce')
plt.title("RozkÅad wartoÅci odstajÄcych dla danych GPU")
plt.show()

# Wizualizacja zmiennych z uwzglÄdnieniem wartoÅci odstajÄcych
plt.figure(figsize=(12, 8))
sns.boxplot(data=df, x='WartoÅci odstajÄce', y='TDP (W)')
plt.title("TDP w podziale na wartoÅci odstajÄce")
plt.show()
```

```python
# 2. RedukowaÄ wymiarowoÅÄ danych z uÅ¼yciem PCA,

numerical_df = df.select_dtypes(include=['float64', 'int64'])

scaler = StandardScaler()
numerical_df_scaled = scaler.fit_transform(numerical_df)

pca = PCA(n_components=2)
pca_result = pca.fit_transform(numerical_df_scaled)

df['PC1'] = pca_result[:, 0]
df['PC2'] = pca_result[:, 1]

pca_df = pd.DataFrame(pca_result, columns=['PC1', 'PC2'])

explained_variance = pca.explained_variance_ratio_
print(f"Wariancja wyjaÅniana przez kaÅ¼dÄ skÅadowÄ: {explained_variance}")
print(f"ÅÄczna wariancja: {sum(explained_variance):.2f}")

plt.figure(figsize=(8, 6))
plt.scatter(pca_df['PC1'], pca_df['PC2'], alpha=0.7)
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('Wizualizacja danych GPU po PCA')
plt.show()
```
